# Evaluator

```python
def evaluator(y_pred, y_label):
    return max(0, 100-sqrt(mean_squared_log_error(y_pred, y_label)))
```

# Lets train

```python
x_train, x_val, y_train, y_val = train_test_split(final_train_df.iloc[:, :-1], final_train_df.iloc[:, -1], random_state = 20, test_size = 0.3)

print(f"Shape of x_train is : {x_train.shape}")
print(f"Shape of y_train is : {y_train.shape}")
print(f"Shape of x_val is : {x_val.shape}")
print(f"Shape of y_val is : {y_val.shape}")
```

```python
xgboost_model = XGBRegressor()

xgboost_model.fit(x_train, y_train)
```

```python
xgboost_y_train_pred = xgboost_model.predict(x_train)
xgboost_y_val_pred = xgboost_model.predict(x_val)
```

```python
print(f"Train score : {evaluator(xgboost_y_train_pred, y_train)}")
print(f"val score : {evaluator(np.abs(xgboost_y_val_pred), y_val)}")
```

# Submission
